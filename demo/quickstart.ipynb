# NumCompute Quickstart (Iris)

This notebook demonstrates the core modules with `demo/iris.csv` and `demo/iris_missing.csv` datasets.

```python
import numpy as np
from numcompute.io import load_csv
from numcompute.sort_search import binary_search, quickselect, topk, stable_sort
from numcompute.rank import percentile, rank
from numcompute.metrics import accuracy, confusion_matrix, f1, mse, precision, recall
from numcompute.preprocessing import StandardScaler, MinMaxScaler, SimpleImputer, OneHotEncoder
from numcompute.pipeline import Pipeline
```

## 1) Load CSV with `io.py`

```python
filename = 'iris.csv'
# species is string -> parsed as NaN in load_csv (expected)
raw_numeric = load_csv(filename, skip_rows=1)
print('Shape:', raw_numeric.shape)
print('NaN count:', int(np.isnan(raw_numeric).sum()))
print('First row:', raw_numeric[0])

# Load the missing data version to show NaN handling in stats/metrics
filename_missing = 'iris_missing.csv'
raw_missing = load_csv(filename_missing, skip_rows=1)
print('Shape (missing):', raw_missing.shape)
print('NaN count (missing):', int(np.isnan(raw_missing).sum()))

# Load missing data with NaN filling to show preprocessing
raw_filled = load_csv(filename_missing, skip_rows=1,
                      fill_value=99.99)  # for demo purposes only, not a real imputation strategy
print('Shape (filled):', raw_filled.shape)
print('NaN count (filled):', int(np.isnan(raw_filled).sum()))
```

## 2) Preprocessing + pipeline

```python


#get only numeric columns (first 4, drop species which is NaN)
X = raw_missing[:, :4]
print("Shape before imputing:", X.shape)
print("NaN count before:", int(np.isnan(X).sum()))


#build a pipeline impute NaN then scale
pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler())
])

X_out = pipe.fit_transform(X)
print("NaN count after:", int(np.isnan(X_out).sum()))
print("Mean per column :", np.round(np.mean(X_out, axis=0), 4))
print("Std per column :", np.round(np.std(X_out, axis=0), 4))
```

```python
#MinMaxScaler 
pipe2 = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", MinMaxScaler(feature_range=(0, 1)))
])

X_minmax = pipe2.fit_transform(X)
print("Min per column :", np.round(np.min(X_minmax, axis=0), 4))
print("Max per column :", np.round(np.max(X_minmax, axis=0), 4))
```
